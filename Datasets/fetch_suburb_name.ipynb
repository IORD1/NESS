```python
for index, row in df.iterrows():
    suburb_name = row['name']
    # Prepare the request URL
    url = f"{geocoding_api_endpoint}?address={suburb_name}+,+Pune&key={api_key}"
    # Send the GET request to Google Geocoding API
    response = requests.get(url)
    data = response.json()
    # Extract latitude and longitude from the response
    if data['status'] == 'OK':
        lat = data['results'][0]['geometry']['location']['lat']
        lon = data['results'][0]['geometry']['location']['lng']
        # Assign the coordinates to the respective suburb
        df.at[index, 'Latitude'] = lat
        df.at[index, 'Longitude'] = lon
    else:
        print(f"Error fetching coordinates for {suburb_name}")

print(df)
```
